```python
import pandas as pd
```

```python
train=pd.read_csv(r"C:\Users\liudonghang\Documents\Python Scripts\Data analysis with pandas\pandas-videos-master\data\titanic_train.csv")
train.head()
```

```python
train['Sex_male'] = train.Sex.map({'male':1, 'female':0})
train[['Sex', 'Sex_male']].head()
```

```python
train.drop('Sex_male', axis=1, inplace=True)
```

```python
Embark_dummies = pd.get_dummies(train.Embarked, prefix='Embark').iloc[:,1:]
```

```python
pd.concat([train, Embark_dummies], axis=1).head()
```

```python
pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True ).head()
```
